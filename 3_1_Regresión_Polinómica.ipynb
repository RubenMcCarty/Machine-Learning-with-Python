# 3. Creando un conjunto de train y test
La divisi칩n de conjunto de Train/Test, implica dividir el dataset en train set y test set, que son mutuamente excluyentes. Despues se debe entrenar con el train set y probar con el test set.

```python
msk = np.random.rand(len(df))<0.8
train = cdf[msk]
test = cdf[~msk]
```

# 4. Regresi칩n Polinomial
A veces, la tendencia de los datos no es realmente lineal y parece curvil칤nea. En este caso podemos utilizar m칠todos de regresi칩n polinomial. De hecho, existen muchas regresiones diferentes que pueden usarse para ajustarse a cualquier aspecto del conjunto de datos, como cuadr치tico, c칰bico, etc.y puede continuar hasta grados infinitos.

En esencia, podemos llamar a todos estos regresi칩n polinomial, donde la relaci칩n entre la variable independiente x y la variable dependiente y se modela como un polinomio de en칠simo grado en x. Digamos que desea tener una regresi칩n polinomial (hagamos un polinomio de 2 grados):

$洧녽 = 洧녪 + \theta_1洧논 + \theta_2洧논^2 $
Ahora, la pregunta es: 쯖칩mo podemos ajustar nuestros datos en esta ecuaci칩n mientras solo tenemos valores x, como el tama침o del motor? Bueno, podemos crear algunas caracter칤sticas adicionales: 1, $洧논$ y $洧논^2$.

La funci칩n **PloynomialFeatures()** en la biblioteca Scikit-learn, impulsa un nuevo conjunto de caracter칤sticas del conjunto de caracter칤sticas original. Es decir, se generar치 una matriz que consta de todas las combinaciones polinomiales de las caracter칤sticas con grado menor o igual al grado especificado. Por ejemplo, digamos que el conjunto de caracter칤sticas original tiene solo una caracter칤stica, ENGINESIZE. Ahora, si seleccionamos el grado del polinomio para que sea 2, entonces genera 3 caracter칤sticas, grado = 0, grado = 1 y grado = 2:

```python
from sklearn.preprocessing import PolynomialFeatures
from sklearn import linear_model
train_x =np.asanyarray(train[['ENGINESIZE']])
train_y = np.asanyarray(train[['CO2EMISSIONS']])

test_x = np.asanyarray(test[['ENGINESIZE']])
test_y = np.asanyarray(test[['CO2EMISSIONS']])

poly = PolynomialFeatures(degree =2)
train_x_poly = poly.fit_transform(train_x)
train_x_poly
```

**fit_transform** toma nuestros valores de x y genera una lista de nuestros datos elevados desde la potencia de 0 a la potencia de 2 (ya que establecemos el grado de nuestro polinomio en 2).

$
\begin{bmatrix}
    v_1\\
    v_2\\
    \vdots\\
    v_n
\end{bmatrix}
$
$\longrightarrow$
$
\begin{bmatrix}
    [ 1 & v_1 & v_1^2]\\
    [ 1 & v_2 & v_2^2]\\
    \vdots & \vdots & \vdots\\
    [ 1 & v_n & v_n^2]
\end{bmatrix}
$

in our example

$
\begin{bmatrix}
    2.\\
    2.4\\
    1.5\\
    \vdots
\end{bmatrix}
$
$\longrightarrow$
$
\begin{bmatrix}
    [ 1 & 2. & 4.]\\
    [ 1 & 2.4 & 5.76]\\
    [ 1 & 1.5 & 2.25]\\
    \vdots & \vdots & \vdots\\
\end{bmatrix}
$

Parecen conjuntos de caracter칤sticas para an치lisis de regresi칩n lineal m칰ltiple, 쯨erdad? Si. Lo hace. De hecho, la regresi칩n polinomial es un caso especial de regresi칩n lineal, con la idea principal de c칩mo seleccionar sus caracter칤sticas. Solo considere reemplazar $洧논$ con $洧논1$, $洧논_1^2$ con $洧논_2$, y as칤 sucesivamente. Entonces la ecuaci칩n de grado 2 se convertir칤a en:

$洧녽 = 洧녪 + \theta_1 洧논_1 +\theta_2 洧논2$

Ahora, podemos tratarlo como un problema de "regresi칩n lineal". Por lo tanto, esta regresi칩n polinomial se considera un caso especial de regresi칩n lineal m칰ltiple tradicional. Entonces, puede usar el mismo mecanismo que la regresi칩n lineal para resolver tales problemas.

as칤 que podemos usar la funci칩n LinearRegression () para resolverlo:

```python
clf = linear_model.LinearRegression()
train_y_ = clf.fit(train_x_poly, train_y)
# The coefficients
print ('Coefficients: ', clf.coef_)
print ('Intercept: ',clf.intercept_)
```

Como se mencion칩 anteriormente, Coeficiente e Intercepci칩n, son los par치metros de la l칤nea curva ajustada. Dado que es una regresi칩n lineal m칰ltiple t칤pica, con 3 par치metros, y sabiendo que los par치metros son la intersecci칩n y los coeficientes del hiperplano, sklearn los ha estimado a partir de nuestro nuevo conjunto de conjuntos de caracter칤sticas. Vamos a trazarlo:

```python
plt.scatter(train.ENGINESIZE, train.CO2EMISSIONS, color = 'blue')
XX = np.arange(0.0,10.0, 0.1)
YY = clf.intercept_[0]+ clf.coef_[0][1]*XX + clf.coef_[0][2]*np.power(XX, 2)
plt.plot(XX, YY, '-r')
plt.xlabel('Tama침o de motor')
plt.ylabel('Emision de CO2')
plt.show
```

# 5.Evaluaci칩n

```python
from sklearn.metrics import r2_score

test_x_poly = poly.fit_transform(test_x)
test_y_ = clf.predict(test_x_poly)

print('MAE: %.2f' % np.mean(np.absolute(test_y_ -test_y)))
print ('MSE: %.2f' % np.mean((test_y_-test_y)**2))
print ('R2_SCORE: %.2f' % r2_score(test_y_, test_y))
```
